```python
import pandas as pd

# Load data
df = pd.read_csv('restaurant_reviews.csv')

# Question 2: Aggregations on Price (encoded from Price Range) grouped by Cuisine Type
price_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['Price'] = df['Price Range'].map(price_map)
result_q2 = df.groupby('Cuisine Type')['Price'].agg(['sum', 'mean', 'count']).reset_index()
result_q2.columns = ['Cuisine Type', 'Total Price', 'Average Price', 'Count']

# Question 3: Min/Max Price grouped by Restaurant Name
result_q3 = df.groupby('Restaurant Name')['Price'].agg(['min', 'max']).reset_index()
result_q3.columns = ['Restaurant Name', 'Min Price', 'Max Price']

# Question 4: Rating range grouped by City (Location)
result_q4 = df.groupby('Location')['Rating'].agg(lambda x: x.max() - x.min()).reset_index()
result_q4.columns = ['City', 'Rating Range']

# Display results
print("Question 2 Results:")
print(result_q2)
print("\nQuestion 3 Results:")
print(result_q3)
print("\nQuestion 4 Results:")
print(result_q4)
```

```python
# Custom aggregation function
def review_range(x):
    return x.max() - x.min()

# Calculate review_count range per city
review_range_by_city = df.groupby('location')['review_count'].agg(review_range)
print("\nReview Count Range per Location:\n", review_range_by_city)
```

```python
import pandas as pd

# Load the dataset
try:
    df = pd.read_csv("restaurant_reviews.csv")  # Replace with your file path
    print("Columns available:", df.columns.tolist())
except FileNotFoundError:
    print("Error: File not found. Check the file path.")
    exit()

# Filter for Rating 2.0-4.0 and Kolkata
filtered_data = df[
    (df["Rating"].between(2.0, 4.0)) &
    (df["Location"] == "Kolkata")
]

# Group by Cuisine Type and aggregate
result = filtered_data.groupby("Cuisine Type", as_index=False).agg(
    Mean_Review_Count=("Review Count", "mean"),
    Median_Review_Count=("Review Count", "median")
)

# Sort by Mean Review Count descending
final_result = result.sort_values("Mean_Review_Count", ascending=False)

print("\nResult:")
print(final_result)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load dataset with error handling
try:
    df = pd.read_csv("restaurant_reviews.csv")
except FileNotFoundError:
    print("Error: File not found. Check path or upload file.")
    exit()

# Check actual columns
print("Original columns:", df.columns.tolist())

# --------------------------------------------------
# 1. Normalize Price Range (instead of Price)
# --------------------------------------------------
if 'Price Range' in df.columns:
    # Convert price categories to numerical values
    price_map = {'Low': 1, 'Medium': 2, 'High': 3}
    df['Price_Numeric'] = df['Price Range'].map(price_map)
    
    # Normalize between 0-1
    scaler = MinMaxScaler()
    df['Price_Normalized'] = scaler.fit_transform(df[['Price_Numeric']])
else:
    print("Price Range column not found - skipping normalization")

# --------------------------------------------------
# 2. Clean Review Count (instead of Age)
# --------------------------------------------------
if 'Review Count' in df.columns:
    # Replace negative review counts (though unlikely) with median
    median_reviews = df['Review Count'].median()
    df['Review Count'] = df['Review Count'].apply(lambda x: x if x >= 0 else median_reviews)
else:
    print("Review Count column not found - skipping cleaning")

# --------------------------------------------------
# 3. Merge with sample product data (demonstration)
# --------------------------------------------------
# Create sample product data (replace with your actual data)
product_data = pd.DataFrame({
    'Restaurant_Name': df['Restaurant Name'].sample(5).tolist(),
    'Popular_Dish': ['Pizza', 'Pasta', 'Biryani', 'Tacos', 'Sushi'],
    'Ingredient_Cost': [4.5, 3.8, 5.2, 3.5, 6.0]
})

# Merge on restaurant name
merged_data = pd.merge(
    df,
    product_data,
    left_on='Restaurant Name',
    right_on='Restaurant_Name',
    how='left'
)

# Show first 5 rows of merged data
print("\nMerged Data (First 5 Rows):")
print(merged_data.head().to_string(index=False))
```

```python
import pandas as pd

# 1. Create original files
# ==============================================
original_data = {
    'Student ID': [101, 102, 103, 104, 105],
    'Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'Diana Prince', 'Evan Wright'],
    'Grade': ['A', 'C', 'B', 'A', 'D'],
    'Score': [85, 42, 67, 92, 38]
}

# Create DataFrame
df = pd.DataFrame(original_data)

# Save original files
df.to_csv('students_original.csv', index=False)
df.to_excel('students_original.xlsx', index=False, engine='openpyxl')

# 2. Process files in Jupyter Notebook
# ==============================================
# Read original files
csv_df = pd.read_csv('students_original.csv')
excel_df = pd.read_excel('students_original.xlsx')

# Add Pass/Fail status to both DataFrames
for dataframe in [csv_df, excel_df]:
    dataframe['Pass/Fail'] = dataframe['Score'].apply(
        lambda x: 'Pass' if x > 50 else 'Fail'
    )

# Save modified files
csv_df.to_csv('students_modified.csv', index=False)
excel_df.to_excel('students_modified.xlsx', index=False, engine='openpyxl')

# 3. Compare files
# ==============================================
print("Original CSV File:")
print(pd.read_csv('students_original.csv').head())

print("\nModified CSV File:")
print(pd.read_csv('students_modified.csv').head())

print("\nOriginal Excel File:")
print(pd.read_excel('students_original.xlsx').head())

print("\nModified Excel File:")
print(pd.read_excel('students_modified.xlsx').head())
```

```python
import pandas as pd

# 1. Create original files with city data
original_data = {
    'City': ['Tokyo', 'Delhi', 'Shanghai', 'São Paulo', 'Mexico City'],
    'Country': ['Japan', 'India', 'China', 'Brazil', 'Mexico'],
    'Population': [37400000, 32000000, 27000000, 22000000, 22000000],
    'Area': [2194, 1484, 6341, 1521, 1485]
}

# Create DataFrame
df = pd.DataFrame(original_data)

# Save original files
df.to_csv('cities_original.csv', index=False)
df.to_excel('cities_original.xlsx', index=False, engine='openpyxl')

# 2. Process files in Jupyter Notebook
# Read original files
csv_df = pd.read_csv('cities_original.csv')
excel_df = pd.read_excel('cities_original.xlsx')

# Add Population Density to both DataFrames
for dataframe in [csv_df, excel_df]:
    dataframe['Population Density'] = dataframe['Population'] / dataframe['Area']

# Save modified files
csv_df.to_csv('cities_modified.csv', index=False)
excel_df.to_excel('cities_modified.xlsx', index=False, engine='openpyxl')

# 3. Compare files
print("Original CSV File:")
print(pd.read_csv('cities_original.csv').head())

print("\nModified CSV File:")
print(pd.read_csv('cities_modified.csv').head())

print("\nOriginal Excel File:")
print(pd.read_excel('cities_original.xlsx').head())

print("\nModified Excel File:")
print(pd.read_excel('cities_modified.xlsx').head())
```
